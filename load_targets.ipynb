```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python

def load_lp(path):
    with open(path) as f:
        _ = f.readline()  # COL_NAMES=...
        row_line = f.readline().strip()
        row_names = row_line.split('=')[1].split('|')[1:]
        data = np.loadtxt(f)            # shape: (4, ncols)
    cols = [f'col{i}' for i in range(data.shape[1])]
    df = pd.DataFrame(data, index=row_names, columns=cols)
    return df

p = Path("jan_26_exp/final_lp_targets/T1.txt")
df = load_lp(p)

all_dfs = []
for i in range(1,16):
    p = Path(f"jan_26_exp/final_lp_targets/T{i}.txt")
    df = load_lp(p)
    all_dfs.append(df)
```

```python
all_dfs[0]
```

## Plot all

```python
prof = 'rot'
ncols = all_dfs[0].shape[1]
x = np.arange(ncols)

for i in range(1,16):
    data = all_dfs[i-1].loc[prof]
    plt.plot(x, data, label=f'T{i}')

# show ticks every 5 indices and label them 1..N accordingly
ticks = np.arange(0, ncols, 5)
labels = [str(i+1) for i in ticks]
plt.xticks(ticks=ticks, labels=labels)
# plt.xlim(0, ncols-1)
plt.xlabel('rad_idx')
plt.grid(True, which='both', linestyle='--', alpha=0.5)
plt.legend()
plt.title(prof)
plt.show()
```
